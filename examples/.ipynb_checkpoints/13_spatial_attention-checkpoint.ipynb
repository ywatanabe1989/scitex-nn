# 13 — `SpatialAttention`: channel-wise gain from a 1×1 conv

Tiny block: average-pool over time, then 1×1 conv `(C → 1)` produces
one scalar per batch which broadcasts back over all channels and
time-points to gate the input. The network learns *how much* of each
channel contributes overall — not where in time it matters.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import torch

import scitex_nn

torch.manual_seed(0)
n_chs, seq_len, fs = 8, 256, 256
t = torch.linspace(0, seq_len / fs, seq_len)
strong = {0, 3, 6}
base = torch.stack(
    [
        torch.sin(2 * torch.pi * (5 + c) * t)
        if c in strong
        else 0.1 * torch.randn(seq_len)
        for c in range(n_chs)
    ],
    dim=0,
)
x = base.unsqueeze(0)
```

## Apply attention

```python
sa = scitex_nn.SpatialAttention(n_chs_in=n_chs)
y = sa(x).detach()
gain = (y / (x + 1e-12)).squeeze(0).mean(dim=-1)
print(f"per-channel gain: {gain.tolist()}")
```

## Input → output → gain

```python
fig, axes = plt.subplots(3, 1, figsize=(9, 5))
for c in range(n_chs):
    axes[0].plot(t.numpy(), x[0, c].numpy() + c * 2.5, lw=0.6)
    axes[1].plot(t.numpy(), y[0, c].numpy() + c * 2.5, lw=0.6)
axes[0].set_title("input — channels 0/3/6 carry signal")
axes[1].set_title("after SpatialAttention")
axes[2].bar(range(n_chs), gain.numpy())
axes[2].set_xlabel("channel")
axes[2].set_ylabel("mean gain")
axes[2].set_title("Per-channel gain")
fig.tight_layout()
plt.show()
```
