# 07 — `Hilbert`: differentiable analytic-signal envelope

Returns the analytic-signal phase and amplitude stacked along a new
last dim. Output matches `scipy.signal.hilbert` to float-32 noise.
Reference: <https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.hilbert.html>.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import torch
from scipy.signal import chirp
from scipy.signal import hilbert as scipy_hilbert

import scitex_nn

fs, duration = 400, 1.0
n = int(fs * duration)
t = np.arange(n) / fs
```

## Build the canonical scipy-doc demo signal: a 20→100 Hz chirp
## ridden by `1 + 0.5·sin(2π·3·t)`.

```python
sig = (chirp(t, 20, t[-1], 100) * (1 + 0.5 * np.sin(2 * np.pi * 3 * t))).astype(
    np.float32
)

H = scitex_nn.Hilbert(seq_len=n, dim=-1)
out = H(torch.from_numpy(sig)).numpy()
amp_stx = out[..., 1]
amp_scipy = np.abs(scipy_hilbert(sig))
diff = float(np.abs(amp_stx - amp_scipy).max())
```

## Overlay scipy and scitex_nn envelopes — they coincide to float-32 noise.

```python
fig, axes = plt.subplots(2, 1, figsize=(9, 4.5), sharex=True)
axes[0].plot(t, sig, "k", lw=0.5, label="signal")
axes[0].plot(
    t, 1 + 0.5 * np.sin(2 * np.pi * 3 * t), "grey", ls=":", label="input envelope"
)
axes[0].legend(loc="upper right")
axes[0].set_title("Chirp 20→100 Hz × (1 + 0.5·sin(2π·3·t))")

axes[1].plot(t, amp_scipy, lw=2, label="scipy.signal.hilbert")
axes[1].plot(t, amp_stx, "--", lw=2, label="scitex_nn.Hilbert")
axes[1].legend(loc="lower right")
axes[1].set_xlabel("time [s]")
axes[1].set_title(f"Recovered envelope — max|diff| = {diff:.2e}")
fig.tight_layout()
plt.show()
assert diff < 1e-5, "Hilbert envelope drifted from scipy"
```
