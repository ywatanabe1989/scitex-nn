# 05 — `PSD`: power spectral density via FFT

`scitex_nn.PSD(sample_rate)` is a one-line wrapper around
`torch.fft.rfft`: it returns `(power, freqs)` for a real-valued
signal. The kernel is batched and runs on GPU; gradients flow.
Reference: `scipy.signal.welch` with `nperseg = seq_len`,
`noverlap = 0` (one Hann window — the simplest periodogram).

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import torch
from scipy.signal import welch

import scitex_nn

rng = np.random.default_rng(0)
fs, seq_len = 500, 2048
t = np.arange(seq_len) / fs


def pink(n):
    spec = np.fft.rfft(rng.standard_normal(n))
    f = np.arange(spec.size, dtype=float)
    f[0] = 1
    return np.fft.irfft(spec / np.sqrt(f), n=n)


signals = {
    "10 Hz sine": np.sin(2 * np.pi * 10 * t),
    "10 + 25 Hz two-tone": np.sin(2 * np.pi * 10 * t)
    + 0.6 * np.sin(2 * np.pi * 25 * t),
    "pink noise (1/f)": pink(seq_len),
}
```

## Three signals, two backends

```python
psd_layer = scitex_nn.PSD(sample_rate=fs)

fig, axes = plt.subplots(len(signals), 1, figsize=(9, 6), sharex=True)
for ax, (name, sig) in zip(axes, signals.items()):
    psd_stx, fr = psd_layer(torch.from_numpy(sig).float())
    f_ref, p_ref = welch(sig, fs=fs, nperseg=seq_len, noverlap=0)
    ax.plot(fr.numpy(), psd_stx.numpy(), lw=1.4, label="scitex_nn.PSD")
    ax.plot(f_ref, p_ref, "--", lw=1.0, label="scipy.welch")
    ax.set_xlim(0, 60)
    ax.set_yscale("log")
    ax.set_ylabel("power")
    ax.set_title(name, fontsize=9)
    ax.legend(loc="upper right")
axes[-1].set_xlabel("frequency [Hz]")
fig.tight_layout()
plt.show()
```
