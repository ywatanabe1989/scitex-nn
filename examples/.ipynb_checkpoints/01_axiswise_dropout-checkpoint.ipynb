# 01 — `AxiswiseDropout`: drop entire slices along an axis

`nn.Dropout` zeros individual elements. `AxiswiseDropout` drops
whole slices along one axis — useful when neighbouring elements in
a slice are correlated (a whole channel, a whole timepoint).

Inputs / outputs share `(batch, channels, samples)` shape.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import torch

import scitex_nn

torch.manual_seed(0)
```

## Channel-wise vs time-wise vs element-wise

```python
x = torch.ones(1, 8, 24)
y_chan = scitex_nn.AxiswiseDropout(dropout_prob=0.5, dim=1).train()(x)
y_time = scitex_nn.AxiswiseDropout(dropout_prob=0.5, dim=2).train()(x)
y_elem = torch.nn.Dropout(p=0.5).train()(x)

fig, axes = plt.subplots(4, 1, figsize=(8, 4), sharex=True)
for ax, m, title in zip(
    axes,
    [x[0], y_chan[0], y_time[0], y_elem[0]],
    [
        "input (all ones)",
        "AxiswiseDropout(dim=1) — channels",
        "AxiswiseDropout(dim=2) — time",
        "nn.Dropout — element-wise",
    ],
):
    ax.imshow(m.detach().numpy(), aspect="auto", cmap="gray_r", vmin=0, vmax=2.5)
    ax.set_ylabel("channel")
    ax.set_title(title, fontsize=9)
axes[-1].set_xlabel("time")
fig.tight_layout()
plt.show()
```

## Eval mode is the identity

```python
drop = scitex_nn.AxiswiseDropout(dropout_prob=0.9, dim=1).eval()
assert torch.allclose(drop(x), x), "AxiswiseDropout should be identity in eval()"
print("eval mode: identity ✓")
```
