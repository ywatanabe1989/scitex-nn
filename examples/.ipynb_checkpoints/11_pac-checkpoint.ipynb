# 11 — `PAC`: end-to-end phase-amplitude coupling pipeline

`scitex_nn.PAC` chains the package's bandpass bank, `Hilbert`, and
`ModulationIndex` into one `nn.Module`. Input `(B, C, T)` gives a
comodulogram indexed by phase-frequency × amplitude-frequency.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import torch

import scitex_nn

rng = np.random.default_rng(0)
fs, duration = 250, 4.0
seq_len = int(duration * fs)
t = np.arange(seq_len) / fs

# Synthetic theta-gamma coupling
theta = np.sin(2 * np.pi * 6 * t)
env = 1 + 0.8 * (theta + 1) / 2
sig = (
    theta + env * np.sin(2 * np.pi * 60 * t) + 0.05 * rng.standard_normal(seq_len)
).astype(np.float32)
x = torch.from_numpy(sig).view(1, 1, -1)
```

## Forward through `PAC`

```python
pac = scitex_nn.PAC(
    seq_len=seq_len,
    fs=fs,
    pha_start_hz=2,
    pha_end_hz=20,
    pha_n_bands=12,
    amp_start_hz=30,
    amp_end_hz=100,
    amp_n_bands=16,
    fp16=False,
)
como = pac(x).squeeze().detach().float().numpy()
pha_mids = pac.PHA_MIDS_HZ.numpy()
amp_mids = pac.AMP_MIDS_HZ.numpy()
peak_pha = pha_mids[int(np.argmax(como.max(axis=1)))]
peak_amp = amp_mids[int(np.argmax(como.max(axis=0)))]
print(f"comodulogram shape: {como.shape}; max MI = {como.max():.4f}")
print(f"peak at ({peak_pha:.1f}, {peak_amp:.1f}) Hz")
```

## Comodulogram

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
n_show = int(1.5 * fs)
axes[0].plot(t[:n_show], sig[:n_show], "k", lw=0.5, label="signal")
axes[0].plot(t[:n_show], theta[:n_show] * 1.5, "grey", ls=":", label="theta 6 Hz")
axes[0].legend(loc="upper right")
axes[0].set_xlabel("time [s]")
axes[0].set_title("Synthetic theta-phase × gamma-amplitude coupling")

im = axes[1].imshow(
    como,
    aspect="auto",
    origin="lower",
    extent=(amp_mids[0], amp_mids[-1], pha_mids[0], pha_mids[-1]),
    cmap="viridis",
)
fig.colorbar(im, ax=axes[1], label="MI")
axes[1].set_xlabel("amplitude freq [Hz]")
axes[1].set_ylabel("phase freq [Hz]")
axes[1].set_title(f"Comodulogram — peak at ({peak_pha:.1f}, {peak_amp:.1f}) Hz")
fig.tight_layout()
plt.show()
```
